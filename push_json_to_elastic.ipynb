```python
# installed ELK (elasticsearch, logstash and kibana using apt-get)
# using data and examples from following git repo: 
# https://github.com/elastic/examples/tree/master/Exploring%20Public%20Datasets/nyc_traffic_accidents
```

```python
import elasticsearch
es = elasticsearch.Elasticsearch()
```

```python
# load and read csv file using pandas
import pandas as pd

# its a very big file (Around 290 mb) it takes some time to load 
csv_file_path = "./nyc_collision_data.csv"
df = pd.read_csv(csv_file_path, header=0)
```

```python
# check if the data got loaded properly
print (df.shape)
df.head()
```

```python
# convert the csv dataframe into json
import json

# clipping the dataframe as it is huge, around 2 million enteries
clipped_df = df[:1000]
data = json.loads(clipped_df.to_json(orient='records'))
```

```python
# push json to ES 
# specify index and type name for elastic search
INDEX = "nyc_collison"
DOC_TYPE = "sample"

# data is a list with each record in the csv (in the form of dictionary) as its element
# iterate through the list and push each record 
# Note: we haven't specified the "id" while pushing the recorde, so they will be assigned randomly by ES
for record in data:
    es.index(index=INDEX, doc_type=DOC_TYPE, body=record)
```

```python
# data is loaded in following format
# http://localhost:9200/[INDEX]/[DOC_TYPE]/

# query the index 
es.search(index="nyc_collison", doc_type="sample", body={"query": {"match": {u'BOROUGH': u'BROOKLYN'}}})
```
